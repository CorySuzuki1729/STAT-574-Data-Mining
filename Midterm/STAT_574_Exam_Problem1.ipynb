```python
# STAT 574 Midterm Exam Problem 1

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVR

hospital_data = pd.read_csv("C:/Users/coryg/OneDrive/Desktop/STAT_574_Data_Mining/hospital_data.csv")
gender_code = {'M':1, 'F':0}
hospital_data["gender"] = hospital_data["gender"].map(gender_code)

X = hospital_data.iloc[:, 0:6].values
y = hospital_data.iloc[:, 6].values

# Splitting the data into 80% training and 20% testing sets. 

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20,
                        random_state=562323)

# Fitting Support Vector Regression with Linear Kernel

svreg_linear = SVR(kernel="linear").fit(X_train, y_train)

# Computing prediction accuracy for testing data. 

y_pred = svreg_linear.predict(X_test)
ind10 = []
ind15 = []
ind20 = []

for sub1, sub2 in zip(y_pred, y_test):
    ind10.append(1) if abs(sub1-sub2)<0.10*sub2 else ind10.append(0)
    ind15.append(1) if abs(sub1-sub2)<0.15*sub2 else ind15.append(0)
    ind20.append(1) if abs(sub1-sub2)<0.20*sub2 else ind20.append(0)

print("")
print("Linear Kernel")

# Accuracy within 10%
accuracy10 = sum(ind10)/len(ind10)
print("within 10%:", round(accuracy10, 5))

# Accuracy within 15%
accuracy15 = sum(ind15)/len(ind15)
print("within 15%:", round(accuracy15, 5))

# Accuracy within 20%
accuracy20 = sum(ind20)/len(ind20)
print("within 20%:", round(accuracy20, 5))

# Fitting support vector regression with Polynomial kernel.

svreg_poly = SVR(kernel="poly").fit(X_train, y_train)

# Computing prediction accuracy on testing data. 
y_pred_poly = svreg_poly.predict(X_test)

ind10_poly = []
ind15_poly = []
ind20_poly = []

for sub1, sub2 in zip(y_pred_poly, y_test):
    ind10_poly.append(1) if abs(sub1-sub2)<0.10*sub2 else ind10_poly.append(0)
    ind15_poly.append(1) if abs(sub1-sub2)<0.15*sub2 else ind15_poly.append(0)
    ind20_poly.append(1) if abs(sub1-sub2)<0.20*sub2 else ind20_poly.append(0)

print("")
print("Polynomial Kernel")
# Accuracy within 10%
accuracy10_poly = sum(ind10_poly)/len(ind10_poly)
print("Within 10%:", round(accuracy10_poly, 5))

# Accuracy within 15%
accuracy15_poly = sum(ind15_poly)/len(ind15_poly)
print("Within 15%:", round(accuracy15_poly, 5))

# Accuracy within 20%
accuracy20_poly = sum(ind20_poly)/len(ind20_poly)
print("Within 20%:", round(accuracy20_poly, 5))

# Fitting support vector regression with radial kernel. 

svreg_radial = SVR(kernel="rbf").fit(X_train, y_train)

# Computing prediction accuracy for testing data. 

y_pred_radial = svreg_radial.predict(X_test)

ind10_radial = []
ind15_radial = []
ind20_radial = []

for sub1, sub2 in zip(y_pred_radial, y_test):
    ind10_radial.append(1) if abs(sub1-sub2)<0.10*sub2 else ind10_radial.append(0)
    ind15_radial.append(1) if abs(sub1-sub2)<0.15*sub2 else ind15_radial.append(0)
    ind20_radial.append(1) if abs(sub1-sub2)<0.20*sub2 else ind20_radial.append(0)

print("")
print("Radial Kernel")
# Accuracy within 10%
accuracy10_radial = sum(ind10_radial)/len(ind10_radial)
print("Within 10%:", round(accuracy10_radial, 5))

# Accuracy within 15%
accuracy15_radial = sum(ind15_radial)/len(ind15_radial)
print("Within 15%:", round(accuracy15_radial,5))

# Accuracy within 20%
accuracy20_radial = sum(ind20_radial)/len(ind20_radial)
print("Within 20%:", round(accuracy20_radial, 5))

# Fitting support vector regression with signmoidal kernel
svreg_sigmoid = SVR(kernel="sigmoid").fit(X_train, y_train)

# Computing prediction accuracy for testing data. 

y_pred_sigmoid = svreg_sigmoid.predict(X_test)

ind10_sigmoid = []
ind15_sigmoid = []
ind20_sigmoid = []

for sub1, sub2 in zip(y_pred_sigmoid, y_test):
    ind10_sigmoid.append(1) if abs(sub1-sub2)<0.10*sub2 else ind10_sigmoid.append(0)
    ind15_sigmoid.append(1) if abs(sub1-sub2)<0.15*sub2 else ind15_sigmoid.append(0)
    ind20_sigmoid.append(1) if abs(sub1-sub2)<0.20*sub2 else ind20_sigmoid.append(0)

print("")
print("Sigmoid Kernel")
# Accuracy within 10%
accuracy10_sig = sum(ind10_sigmoid)/len(ind10_sigmoid)
print("Within 10%:", round(accuracy10_sig, 5))

# Accuracy within 15%
accuracy15_sig = sum(ind15_sigmoid)/len(ind15_sigmoid)
print("Within 15%:", round(accuracy15_sig, 5))

# Accuracy within 20%
accuracy20_sig = sum(ind20_sigmoid)/len(ind20_sigmoid)
print("Within 20%:", round(accuracy20_sig, 5))
```
