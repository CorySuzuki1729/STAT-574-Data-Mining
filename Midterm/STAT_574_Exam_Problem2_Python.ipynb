```python
# STAT 574 Midterm Exam Problem 2
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC

card_data = pd.read_csv("C:/Users/coryg/OneDrive/Desktop/STAT_574_Data_Mining/card_transdata.csv")
X = card_data.iloc[:, 0:7].values 
y = card_data.iloc[:, 7].values

# Splitting data into 80% training and 20% testing sets. 

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20,
                            random_state=909808)

# Fitting support vector binary classifier with linear kernel. 

svc_lin = SVC(kernel="linear").fit(X_train, y_train)

# Computing prediction accuracy for testing data. 

y_pred = svc_lin.predict(X_test)
y_test = pd.DataFrame(y_test, columns=["fraud"])
y_pred = pd.DataFrame(y_pred, columns=["predicted"])
df = pd.concat([y_test, y_pred], axis=1)

match = []
for i in range(len(df)):
    if df["fraud"][i] == df["predicted"][i]:
        match.append(1)
    else:
        match.append(0)

accuracy = sum(match)/len(match)

print("")
print("Linear Kernel")
print("Percent accuracy:", accuracy)

# Fitting support vector binary classifier with polynomial kernel. 

svc_poly = SVC(kernel="poly").fit(X_train, y_train)

# Computing prediction accuracy for testing data. 

y_pred = svc_poly.predict(X_test)
y_test = pd.DataFrame(y_test, columns=["fraud"])
y_pred = pd.DataFrame(y_pred, columns=["predicted"])
df = pd.concat([y_test, y_pred], axis=1)

match2 = []
for i in range(len(df)):
    if df["fraud"][i] == df["predicted"][i]:
        match2.append(1)
    else:
        match2.append(0)

accuracy2 = sum(match2)/len(match2)

print("")
print("Polynomial Kernel")
print("Percent Accuracy:", accuracy2)

# Fitting support vector binary classifier with radial kernel. 

svc_radial = SVC(kernel="rbf").fit(X_train, y_train)

# Computing prediction accuracy for testing data. 

y_pred = svc_radial.predict(X_test)
y_test = pd.DataFrame(y_test, columns=["fraud"])
y_pred = pd.DataFrame(y_pred, columns=["predicted"])
df = pd.concat([y_test, y_pred], axis=1)

match3 = []
for i in range(len(df)):
    if df["fraud"][i] == df["predicted"][i]:
        match3.append(1)
    else:
        match3.append(0)

accuracy3 = sum(match3)/len(match3)

print("")
print("Radial Kernel")
print("Percent accuracy:", accuracy3)

# Fitting support vector binary classifier with sigmoid kernel.

svc_sigmoid = SVC(kernel="sigmoid").fit(X_train, y_train)

# Computing prediction accuracy on testing data. 

y_pred = svc_sigmoid.predict(X_test)
y_test = pd.DataFrame(y_test, columns=["fraud"])
y_pred = pd.DataFrame(y_pred, columns=["predicted"])
df = pd.concat([y_test, y_pred], axis=1)

match4 = []
for i in range(len(df)):
    if df["fraud"][i] == df["predicted"][i]:
        match4.append(1)
    else:
        match4.append(0)

accuracy4 = sum(match4)/len(match4)
print("")
print("Sigmoid Kernel")
print("Percent accuracy:", accuracy4)
```
